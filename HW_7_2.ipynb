```r
cox_model <- coxph(Surv(time,status)~age + rx, data = colon_ds)
summary(cox_model)
```

Лечение Lev+5FU значимо снижает риск смерти (HR = 0.657, p = 7e-07)

```r
cox_model2 <- coxph(Surv(time,status)~surg + rx + node4, data = colon_ds)
summary(cox_model2)
```

Операция (surg) значимо увеличивает риск смерти (HR = 1.277, p = 0.000701).
Лечение Lev+5FU (rxLev+5FU) значимо снижает риск смерти (HR = 0.662, p = 1.11e-06).
Отсутствие лечения (rxObs) не оказывает значимого влияния на выживаемость (HR = 1.013, p = 0.864).
Наличие метастазов в 4 или более лимфоузлах (node4) значимо увеличивает риск смерти (HR = 2.486, p < 2e-16)
